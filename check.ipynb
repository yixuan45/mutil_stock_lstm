# 转换为5分钟数据

```python
df_5m = data.copy()
df_5m['t'] = (df_5m['t'] // 300000) * 300000
df_5m = df_5m.groupby('t').agg({
    'o': 'first',
    'h': 'max',
    'l': 'min',
    'c': 'last',
    'v': 'sum',
    'qv': 'sum'
}).reset_index()
df_5m['t'] = df_5m['t'] + 60000 + 5 * 60000
```

# 转化为1个小时数据

```python
df_1h = data.copy()
df_1h['t'] = (df_1h['t'] // 3600000) * 3600000
df_1h = df_1h.groupby('t').agg({
    'o': 'first',
    'h': 'max',
    'l': 'min',
    'c': 'last',
    'v': 'sum',
    'qv': 'sum'
}).reset_index()
df_1h['t'] = df_1h['t'] + 60000 + 60 * 60000
```

# 获取从一月一号到现在的数据

```python
def get_raw_data(path):
    base_url = path['filepath']
    path1 = f"/{path['symbol']}/{path['interval']}"
    url = base_url + path1
    filelist = os.listdir(url)
    filelist.sort()
    filelist_get = [item for item in filelist if
                    path['start_date'] <= item[len(path['symbol']) + 4:len(path['symbol']) + 14] <= path[
                        'end_date']]
    df = pd.DataFrame()
    for file in tqdm(filelist_get):
        df0 = pd.read_csv(fr'{url}/{file}')
        df = pd.concat([df, df0])
    return df.copy()
```

```python
paras1 = {
    'symbol': 'ETHUSDT',
    'start_date': '2025-01-01',
    'end_date': '2025-10-27',
    'interval': '1m',
}
```

```python
# 1. klines data get
path_params1 = {
    'filepath': 'Z:\\um\daily\klines',
    'symbol': paras1['symbol'],
    'start_date': paras1['start_date'],
    'end_date': paras1['end_date'],
    'interval': paras1['interval'],
}
df = get_raw_data(path_params1)
df_total_data = df[['open_time', 'open', 'high', 'low', 'close', 'volume', 'quote_volume']].copy()
df_total_data.columns = ['t', 'o', 'h', 'l', 'c', 'v', 'qv']
df_total_data
```

```python
df_total_data.to_csv("./data/total_data.csv", index=False)
```
